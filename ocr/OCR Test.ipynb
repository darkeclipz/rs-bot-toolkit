```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open('screenshot.png')
```

```python
plt.imshow(image)
```

```python
S = np.array(image).astype('uint8')
```

```python
plt.imshow(S)
```

```python
text_height, text_width = 28, 600
s = S[:text_height, :text_width, :]
plt.imshow(s)
plt.axis('off')
```
